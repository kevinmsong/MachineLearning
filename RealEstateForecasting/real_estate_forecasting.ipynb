```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.externals import joblib
from sklearn.metrics import mean_squared_error
```

```python
np.random.seed(0)
```

```python
data = pd.read_csv("validation.csv.txt", low_memory = False)
```

```python
data = data.drop(["parcel"], axis = 1)
```

```python
data
```

```python
fcv_21 = data["fcv_21"]
```

```python
fcv_19 = pd.DataFrame(data["fcv_20"].tolist())
```

```python
fcv_19.columns = ["fcv_19"]
```

```python
X = data.drop(["fcv_19", "fcv_20", "fcv_21"], axis = 1)
```

```python
X = pd.concat([X, fcv_19], axis = 1)
```

```python
X
```

```python
model = joblib.load("xgb_best.pkl")
```

```python
predictions = model.predict(X)
```

```python
np.sqrt(mean_squared_error(fcv_21, predictions))
```
