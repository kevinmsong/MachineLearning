```python
import numpy as np
import pandas as pd
import xgboost as xgb
```

```python
data = pd.read_csv("PresidentialApproval.csv")
```

```python
data
```

```python
y_train = data.iloc[5:,].Two_or_more_terms
X_train = data.iloc[5:,].drop(columns = "Two_or_more_terms")
```

```python
y_val = data.iloc[1:5,].Two_or_more_terms
X_val = data.iloc[1:5,].drop(columns = "Two_or_more_terms")
```

```python
X_test = pd.DataFrame(data.iloc[0,].drop(columns = "Two_or_more_terms")).T
```

```python
X_test = X_test.drop(columns = "Two_or_more_terms")
```

```python
model = xgb.XGBClassifier()
```

```python
eval_set = [(X_val, y_val)]
```

```python
model.fit(X_train, y_train, early_stopping_rounds=20, eval_set=eval_set, verbose=True)
```

```python
model.predict_proba(X_test)
```
