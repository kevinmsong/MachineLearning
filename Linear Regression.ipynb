## calculating linear regression (with computing)

```python
x_bar = np.mean(x)
y_bar = np.mean(y)
xx_bar = np.mean(x ** 2)
xy_bar = np.mean(x * y)
```

```python
beta_1 = (x_bar * y_bar - xy_bar) / (x_bar ** 2 - xx_bar)
```

```python
beta_0 = (y_bar - x_bar * ((x_bar * y_bar) - (xy_bar))/ (x_bar ** 2 - xx_bar))
```

```python
y_hat = beta_0 + beta_1 * x
```

```python
plt.scatter(x, y)
plt.scatter(x, y_hat)
plt.scatter(x_bar, y_bar)
```

## linear regression class

```python
class LinearRegression:
    def __init__(self):
        self.beta_0 = 0
        self.beta_1 = 0
    
    def fit(self, x, y):
        x_bar = np.mean(x)
        y_bar = np.mean(y)
        xx_bar = np.mean(x ** 2)
        xy_bar = np.mean(x * y)
        
        self.beta_1 = (x_bar * y_bar - xy_bar) / (x_bar ** 2 - xx_bar)
        self.beta_0 = (y_bar - x_bar * ((x_bar * y_bar) - (xy_bar))/ (x_bar ** 2 - xx_bar))
        
    def predict(self, x):
        y_hat = self.beta_0 + self.beta_1 * x
        return y_hat
```

```python
%%time

linear_regressor = LinearRegression()

linear_regressor.fit(x, y)

y_hat_from_lr_class = linear_regressor.predict(x)

plt.scatter(x, y)
plt.scatter(x, y_hat_from_lr_class)
plt.scatter(x_bar, y_bar)
```
